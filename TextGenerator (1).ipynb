```python
# markov chains

def load_text(filename):
    with open(filename,encoding='utf8') as f:
        return f.read().lower()
```

```python
text=load_text(dataset_file)
```

```python
print(text[:500])
```

```python
def genTable(data,k=4):
    T = {}
    for i in range(len(data)-k):
        X=data[i:i+k]
        Y=data[i+k]
        print(X,Y)
        
        if(T.get(X)) is None:
            T[X] = {}
            T[X][Y]=1
        else:
            if T[X].get(Y) is None:
                T[X][Y]=1
            else:
                T[X][Y]+=1
    return T
```

```python
def convertFreqIntoProb(T):
    for kx in T.keys():
        s=sum(T[kx].values())
        for k in T[kx].keys():
            T[kx][k]= T[kx][k]/s
    return T
```

```python
convertFreqIntoProb(T)
```

```python
def train(data,k):
    T=genTable(data,k)
    T=convertFreqIntoProb(T)
    return T
```

```python
T=train(text,4)
```

```python
words=["apple","mango","banana"]
prob=[.2,.7,.1]

import numpy as np
```

```python
# sampling from a probability distriution
for i in range(10):
    print(np.random.choice(words,p=prob))
```

```python
def sample_next(ctx,T):
    ctx=ctx[-4:]
    if T.get(ctx) is None:
        return " "
    possible_chars=list(T[ctx].keys())
    possible_values=list(T[ctx].values())
#     print(possible_chars)
#     print(possible_values)

    return np.random.choice(possible_chars,p=possible_values)
```

```python
sample_next("coun",T)
```

```python
def generateText(starting_sent,maxLen=1000):
    np.random.seed(1)
    sentence=starting_sent
    ctx=starting_sent[-4:]
    
    for ix in range(maxLen):
        next_prediction=sample_next(ctx,T)
        sentence+=next_prediction
        ctx=sentence[-4:]
    return sentence
```
